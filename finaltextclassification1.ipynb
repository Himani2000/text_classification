```python
#importing the necessory modules
import os
import string
import email
import numpy as np
import pandas as pd
```

```python
#giving the paths to the variable 
TestFilePath = './test'
StockFile = './stopwords_en.txt'
FilePath = './newsgroupstrain'
cnt_words = {}
y_pred=[]
```

```python
# here the code is keeping the count of all the words excluding the email header , punctuations and the digits
# and storing them in a dictionary which has the keys as the name of  the word and the values as the count 
loldict={}
i=-1
with open(StockFile) as f:
    stockWords = f.read()
    stockWords = stockWords.split('\n')
for (curdir, subsHere, filesHere) in os.walk(FilePath):
    i=i+1
    for fileHere in filesHere:
        loldict[fileHere]=i
        for line in open(os.path.join(curdir,fileHere)):
            line = line.strip().lower()
            b = email.message_from_string(line)
            b=b.get_payload()
            #b=b.translate(string.punctuation)
            b = b.translate(str.maketrans('','',string.punctuation))
            b = b.translate(str.maketrans('','','1234567890'))
            b = b.split()
            for word in b:
                if word not in stockWords:
                    if word not in cnt_words :
                        cnt_words[word] = 1
                    else:
                        cnt_words[word] += 1
```

```python
def fit(data):# a dataframeis passed 
    output_name = data.columns[-1] # the values are splitted 
    features = data.columns[0:-1]
    counts = {}
    counts["total_count"] = len(data)
    possible_outputs = set(data[output_name])
    for output in possible_outputs:
        counts[output] = {}
        smallData = data[data[output_name] == output]
        counts[output]["total_count"] = len(smallData)
        for i in range(len(features)):
            f = features[i]
            counts[output][i] = {}
            val_count=0
            val_count = len(smallData[smallData[f] == output])
            counts[output][i] = val_count   
    return counts      
```

```python
def fit2(data): 
    output_name = data.columns[-1]
    features = data.columns[0:-1]
    counts = {}
    cp=0
    possible_outputs = set(data[output_name])
    classes=set(y_train)
    counts["total_count"]=len(y_train)
    for current_class in classes:
        counts[current_class]={}
        current_class_rows=(y_train==current_class)
        x_currentrows=x_train[current_class_rows]
        y_currentrows=y_train[current_class_rows]
        features=x_train.shape[1]
        counts[current_class]["total_count"]=len(y_currentrows)
        j=1
        for j in range(features+1):
            counts[current_class][j]=x_currentrows[:,j-1].sum() 
            cp=cp+x_currentrows[:,j-1].sum() 
        counts[current_class]["count"]=cp
        cp=0 
    return counts
```

```python
def predict(dictionary,X_test):
    ypred=[]
    for x in X_test:
        x_class=predictSinglePoint(dictionary,x) 
        ypred.append(x_class) 
    return ypred
```

```python
def predictSinglePoint(dictionary,x_test):
    classes=dictionary.keys()
    best_p=-1000
    best_class=-1
    first_run=True
    for current_class in classes:
        if(current_class=="total_count"):
            continue
        p_current_class=probability(dictionary,x_test,current_class)
        if(first_run or p_current_class>best_p):
            best_p=p_current_class
            best_class=current_class
        first_run=False
    return best_class
```

```python
def probability(dictionary,x,current_class):
    output=0
    features=len(dictionary[current_class].keys())-3
    #print(features)
    for j in range(1,features+1):
        if(x[j-1]==0):
            a=1#  if the data is not on training but is on the tesing so by using laplace correction giving some small probability to it 
            b=features
            current_xjprob=np.log(a)-np.log(b)        
        else:
            a=dictionary[current_class][j]+1  # basically the probability  is calculated by  the prsence of the given word for a specific class
            b=dictionary[current_class]["count"]+features #  also adding the lapalce correction
            current_xjprob=np.log(a)-np.log(b) 
        
        output=output+current_xjprob 
    count_class = dictionary[current_class]["total_count"]
    total_count = dictionary["total_count"]
    class_prob = np.log(count_class) - np.log(total_count)
    output = output + class_prob
    return output
```
